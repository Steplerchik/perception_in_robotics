## Task B. Pre-multiplying matrices
### The code

```python
import tools.jacobian as j
import tools.task as t
import numpy as np
from field_map import FieldMap
```

```python
u = np.array([0, 10, 0])
x = [180, 50, 0]
alphas = np.array([0.05**2, 0.001**2, 0.05**2, 0.01**2])

_, V = j.state_jacobian(x ,u)
print('V:\n', V, '\n')
M = t.get_motion_noise_covariance(u, alphas)

print('M:\n', M, '\n')

print('Pre-multiplying matrice for transition:\n', np.linalg.cholesky(np.linalg.inv(V.dot(M.dot(V.T)))).T, '\n')

Q = np.array([[100, 0], [0, 100]])
print('Pre-multiplying matrice for observation:\n', np.linalg.cholesky(np.linalg.inv(Q)).T, '\n')
```
